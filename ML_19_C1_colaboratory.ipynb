# <h1>Tutorial Google Colab</h1>
Colaboratory es un entorno gratuito de Jupyter Notebook que no requiere configuración y que se ejecuta completamente en la nube.

Colaboratory permite escribir y ejecutar código, guardar y compartir tus análisis y tener acceso a recursos informáticos muy potentes, todo de forma gratuita desde el navegador.


> Todos los cuadernos de Colaboratory se almacenan en Google Drive. Estos se pueden compartir de la misma manera que las Hojas de cálculo o los Documentos de Google. Simplemente haz clic en el botón para compartir que aparece en la parte superior derecha de los cuadernos de Colaboratory o sigue estas instrucciones para compartir archivos en Google Drive.

> Si eliges compartir un cuaderno, se comparte todo su contenido (texto, código y resultado). Para impedir que se guarde o comparta el resultado de la celda de código, selecciona Editar > Configuración del cuaderno > Omitir el resultado de la celda al guardar este cuaderno. De este modo, no se compartirá la máquina virtual que estés usando, incluidos los archivos y las bibliotecas personalizados que hayas configurado. Por lo tanto, es buena idea incluir celdas que instalen y carguen todas las [bibliotecas](https://colab.research.google.com/notebooks/snippets/importing_libraries.ipynb) o los [archivos](https://colab.research.google.com/notebooks/io.ipynb) personalizados que necesite tu cuaderno.

> Todos los cuadernos de Colaboratory se almacenan en el formato de cuaderno de Jupyter de código abierto (.ipynb).
Para descargar los cuadernos de Colaboratory que hayas creado en Google Drive,  accede al menú "Archivo" de Colaboratory.

> Una opción bastante interesante de Colab es la posibilidad de guardar y fijar revisiones de los cuadernos, restaurarlos e incluso descargarlos. Desde el menú "Archivo"

## Utilizar ficheros en local

Subir ficheros desde el sistema local






```
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```
